### Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

## Quick Intro



1. Pandas is quite a game changer when it comes to analyzing data with Python and it is one of the most preferred and widely used tools in data munging/wrangling if not THE most used one.


2. Pandas  takes data (like a CSV or TSV file, or a SQL database) and creates a Python object with rows and columns called data frame that looks very similar to table in a statistical software.


3. This is so much easier to work with in comparison to working with lists and/or dictionaries through for loops or list comprehension.

### The pd is a very popular alias given to pandas

```python
import pandas as pd
```

# Working with Pandas

When you want to use Pandas for data analysis, you’ll usually use it in one of three different ways:

  1. Convert a Python’s list, dictionary or Numpy array to a Pandas data frame
   
   
  2. Open a local file using Pandas, usually a CSV file, but could also be a delimited text file (like TSV), Excel, etc
  
  
 3. Open a remote file or database like a CSV or a JSONon a website through a URL or read from a SQL table/database

# Datastructure in Pandas

There are two types of data structures in pandas: DataFrames and Series.
### DataFrame:
A pandas DataFrame is a two (or more) dimensional data structure – basically a table with rows and columns. The columns have names and the rows have indexes.


```python
df = pd.read_csv('developer_survey_2019/survey_results_public.csv')
```

```python
df
```

### Series
A pandas Series is a one dimensional data structure  that can store values — and for every value it holds a unique index, too.

```python
series = df['Country']
```

```python
series
```

# Viewing And Inspecting Data

Now that you’ve loaded your data, it’s time to take a look. How does the data frame look? Running the name of the data frame would give you the entire table.

You can view the first n rows using df.head(n)

By default, n=5.


```python
df.head()
```

```python
df.head(3)
```

#  .

Similarly, you can view the last n rows using df.tail(n)

Again,by default, n=5.

```python
df.tail()
```

```python
df.tail(2)
```

# .
You can also view n random rows of the data frame using sample method.
##### Note: Here the default value of n is not 5, it is 1.

```python
df.sample()
```

```python
df.sample(5)
```

### Shape
df.shape() gives you the number of rows and columns of the dataframe respectively.

##### Note: df.shape is attribute not function so there won't be any parenthesis at the end.

```python
df.shape
```

### Info
Info Method  gives you the index, datatype and memory information.
It also tells you about the number of rows and columns.

##### PS: This one is a method not an attribute. 

```python
df.info()
```
