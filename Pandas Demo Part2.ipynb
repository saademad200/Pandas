## Selecting rows from DataFrame


```python
df.iloc[0]
```

###### Now for multiple rows


```python
df.iloc[[0,1]]
```

### Now we will filter rows and columns simultaneously

The first arg inside iloc function will be used to filter rows and second will be used to filter columns


```python
df.iloc[[0,1,2],6]
```

```python
df.iloc [[0,1,2],[6,7,8]]
```

### Now we will do the same thing using loc instead of iloc
We will see how they are different

```python
df.loc [0]
```

```python
df.loc [[0,1]]
```

```python
df.loc[[0,1],0]
```

### Wait ✋✋✋✋✋✋

What just happened, why did we get an error.
It's because unlike iloc, loc takes the column names as argument.

Same is the case with row, but right now, our indices numerical values so it works the same as iloc. When we will learn how to change index, after a while longer, we will practice this feature again.

# Indexes 
# ✴️✴️✴️✴️

Default Indexes are just a range of numbers that are basically the integer identifier for the rows.
Sometimes it might make more sense to use a differnet identifier for the rows.


Pandas does not enforce indexes to be unique but it is a good practice to use unique identifiers.

let's create a mini_df and we will be using "Respondent" as "Index"



```python
mini_df = df.loc[0:10,['Respondent','Student', 'Country']]
```

```python
mini_df
```

### As you can see Respondent is unique for each row, we will set it as the index

```python
mini_df.set_index("Respondent")
```

### pssst ' Try typing mini_df in a cell and run it'
# 🙊  🙊  🙊  🙊

```python
mini_df
```

### What? This can't be right!!!

# 🤯  🤯   🤯   🤯   🤯

We just changed the index...... but......

Actually, Pandas do not usually do these changes inplace, unless we specifically tell it to do so.

### Pandas you evil monster

#  😈   😈   😈   😈   😈   

Well, it is actually nice because it allows us to experiment with the data without worrying about modifying our data.

```python
mini_df.set_index("Respondent",inplace=True)
```

```python
mini_df
```
