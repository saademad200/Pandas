# Sorting Data

<img src="img/sort.jpeg" width="250" height="250" />

```python
import pandas as pd
df = pd.read_csv('developer_survey_2019/survey_results_public.csv')
pd.set_option('max_rows',8,'max_columns',14)
df
```

### Let's call our trusty mini_df

<img src="img/call.jpg" width="250" height="250" />

```python
mini_df=df.loc[0:7,['Age','Country','Gender','Student']]
```

```python
mini_df
```

### Let's start sorting

```python
mini_df.sort_values(by='Age')
```

```python
mini_df.sort_values(by='Age',ascending=False)
```

### Now we will sort using two arguments. First argument has higher priority.

```python
mini_df.sort_values(by=['Age','Country'])
```

### The sorting order for both the arguments may not be same

```python
mini_df.sort_values(by=['Age','Country'], ascending=[False,True])
```

```python
mini_df
```

```python
mini_df.sort_values(by=['Age','Country'], ascending=[False,True],inplace=True)
```

```python
mini_df
```
