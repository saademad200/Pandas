### Next up is groupby()
groupby() function is used to split the data into groups based on some criteria.  The abstract definition of grouping is to provide a mapping of labels to group names.


```python
df.groupby(['Country'])
```

```python
country_grp = df.groupby(['Country'])
```

# Task: What the percentage of people who know python from each country
### Total Number of respondents for each country

```python
country_res = df['Country'].value_counts()
country_res
```

### Total number of people who know python

```python
country_py = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_py
```

### Create a new dataframe containing two columns, first number of respondents in each country and second the number of python users in each country
#### Note: Index of this df will be column name

```python
new_df = pd.concat([country_res,country_py],axis='columns')
```

```python
new_df
```

### Let's rename our columns

```python
new_df.rename(columns = {'Country':'NumRespondents','LanguageWorkedWith':'NumPython'},inplace= True)
```

### Now we will create our new column, percentage of people knowing python
Formula => ((Total Number of People Knowing Python)/(Total Number of Respondents)) * 100

```python
new_df['PercentKnowPython'] = (new_df['NumPython']/new_df['NumRespondents']) *100
```

```python
new_df
```

### Now we will sort our data in descending order, with the country having individulas highest percentage of python knowing individuals at the top

```python
new_df.sort_values(by = 'PercentKnowPython',ascending=False,inplace=True)
```

```python
new_df
```

### You see sometimes data can be misleading. for e.g Timor-Leste has only one respondent and he knows python.. so the data shows that 100% Timor-Leste know python..but this isn't necessarily true.

### Now we will see the data of countries with over 50 respondents

```python
new_df[(new_df['NumRespondents']>50)]
```

```python
new_df.loc['China']
```
