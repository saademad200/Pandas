### We will be using four functions i.e apply, map, applymap and replace....All of these functions have very subtle difference between them, so I require your utmost concentration here. 


### Apply function behaves differently with series and dataframe.
### Format of apply function is something like this" series.apply(function) "...note: function will be without parenthesis inside apply function.
### First let's see how it works with Series.

# 🙈   🙈   🙈   🙈   🙈   🙈


```python
mini_df['MAIN_BRANCH'].apply(len)
```

### We just saw how apply works with series using a builtin function. 
### Now let's do it with user defined function. For that we will make a function makelower.

```python
def makelower(hobby):
    return hobby.lower()
```

```python
mini_df['HOBBYIST'] = mini_df['HOBBYIST'].apply(makelower)
```

```python
mini_df
```

### It worked the same way as with builtin function.
#  🤯   🤯   🤯   🤯   🤯 
### Well Duh!!!

### We will now do the same thing with lambda function. Those of you who know lambda function, Kudos!  Those of you who don't know it, don't worry you can survive through this tutorial without it. But it would be good if you checked it out.

```python
mini_df['HOBBYIST'] = mini_df['HOBBYIST'].apply(lambda x: x.upper())
```

```python
mini_df
```

### Now let's see how apply works with dataframe(we discussed previously that it behave differently with dataframe).
# 👀   👀   👀   👀   👀

```python
df.apply(len)
```

### Let's review same concepts
### When we used series.apply(len), what happenned? The length of every single entry of the series was output.
### Incase of dataframe.apply(len), the number of rows in every single column is displayed.
### Now we will see 
1. len(dataframe)
2. len (series)

```python
len(df)
```

```python
len(df['Country'])
```

### len(dataframe) gave the number of rows in the data frame.
### len(series) gave the number of rows in series.
### So technically, len(dataframe) and len(series) do the same thing.
# 🔎   🔎   🔎   🔎   🔎   🔎

### Bear with me a bit longer, the lecture is about to end.
### In our mini_df, ENTRY_NUMBER column has integer entries, I will convert those integer entries to strings, so that we can smoothly run our next function.

```python
mini_df['ENTRY_NUMBER'] = mini_df['ENTRY_NUMBER'].apply(lambda x:str(x))
```

```python
mini_df
```

### Applymap will apply the function given to every entry in the dataframe.
### Note: Applymap does the same thing to dataframe, what apply does to series.

```python
mini_df.applymap(len)
```

```python
mini_df.applymap(str.lower)
```

### PS applymap(like apply) does not amend the original dataframe.

```python
mini_df
```

### The map() function is used to map values of Series according to input correspondence. Used for substituting each value in a Series with another value.

```python
mini_df['HOBBYIST'].map({'NO':'YES'})
```

### Wherever there was 'NO', map function changed it to 'YES'. And for  the entries which were not previously 'NO', NaN is returned.

### Another thing, both map and replace(which we will be discussing after map), do not change the original data.

```python
mini_df
```

```python
mini_df['HOBBYIST'].replace({'NO':'YES'})
```

### I hope you have noticed the difference between map and replace.
# 🕊️   🕊️   🕊️   🕊️   🕊️   🕊️
### Replace does not return NaN, instead it returns the pre-existing entry.

```python
mini_df['HOBBYIST'] = mini_df['HOBBYIST'].replace({'NO':'YES'})
```

```python
mini_df
```
