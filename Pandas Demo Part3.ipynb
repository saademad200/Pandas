```python
import pandas as pd
df = pd.read_csv('developer_survey_2019/survey_results_public.csv')
pd.set_option('max_rows',6)
df
```

# Filtering Data

# 🚿   🚿   🚿  🚿

### df['column_name'] == some_value 
What do you think this will return

# 🧐   🧐   🧐   🧐   🧐

No, it won't return a dataframe with rows that have 'some_value' in corresponding 'column_name'.

It will return a Series of boolean Values.

```python
df['OpenSourcer'] == 'Never'
```

#### Note: We will place this statement in a variable....We can not name variables identical to a builtin function, so we won't name it filter, we will name the variable filt.

# 🎆   🎆   🎆   🎆    🎆

```python
filt = (df['OpenSourcer'] == 'Never')
```


#  🎗️   🎗️   🎗️   🎗️   🎗️

Remember, series is a unidimensional array.
When we place this boolean series inside the square brackets i.e df['series'],
the resultant dataframe will contain rows corresponding to the True value and it won't contain rows corresponding to False values.




```python
df[filt]
```

Although, in this case, df[filt] & df.loc [filt] works the same way, it is better practice to use df.loc[filt]

#  🛸   🛸   🛸   🛸   🛸

```python
df.loc [filt]
```

## Let's Play with these functions a bit to get  better understanding

# 🤾‍  🤾‍   🤾‍  🤾‍  🤾‍

```python
df.loc[filt,'Respondent':'Student']
```

We Can use boolean algebra to place multiple conditions in the same filter

#   🅱️   🅱️   🅱️   🅱️   🅱️

```python
filt = (df['OpenSourcer'] == 'Never') & (df['Hobbyist'] == 'Yes')
```

```python
df.loc[filt,['Employment','OpenSourcer','Hobbyist','LanguageWorkedWith']]
```

```python
filt = (df['OpenSourcer'] == 'Never') | (df['Hobbyist'] == 'Yes')
```

```python
df.loc[filt,['Employment','OpenSourcer','Hobbyist','LanguageWorkedWith']]

```

```python
df.loc[~filt,['Employment','OpenSourcer','Hobbyist','LanguageWorkedWith']]
```

We will create a filter called OverAge, for segregating the data of people over 17.

```python
OverAge = (df['Age'] > 17)
```

```python
df.loc[OverAge,['Age','UndergradMajor','Employment','Student','Country']]
```

### Now we will be doing a newer and very useful concept, so your utmost concentration is required

#  🤓   🤓   🤓   🤓   🤓

We our going to make a list of a few countries and we are going to use isin() function to filter the data of only these countries

```python
countries = ['Pakistan','India','Bangladesh']
```

```python
filt = df['Country'].isin(countries)
```

```python
df.loc[filt,['Age','UndergradMajor','Employment','Student','Country']]
```

### Let's filter out a little more specific data.

#  👨‍💻    👨‍💻    👨‍💻    👨‍💻    👨‍💻

```python
filt = df['Country'].isin(countries) & (df['Age'] >= 18) & (df['Student'] == 'Yes, full-time')
```

```python
df.loc[filt,['Age','UndergradMajor','Employment','Student','Country']]
```

### We will first use str attribute of Pandas Series 
### Then we will use contains function to check whether that particular string is present in the entry.

#  💦   💦   💦   💦   💦

```python
filt = df ['LanguageWorkedWith'].str.contains('Python')
```

```python
filt
```

#  🧿   🧿   🧿   🧿   🧿

Our next step i.e df [filt] is probably not going to work.

No, it's absolutely not gonna work 

Have you guessed why 

You have not

Well that's a shame 

#  😓  😓  😓  😓  😓

It won't because this series is not all boolean ... it has NaN entries, which will cause error.



```python
df[filt]
```

### Here is how we will remove the NaN entries

#  👇  👇  👇  👇  👇

```python
filt = df ['LanguageWorkedWith'].str.contains('Python', na = False)
```

```python
df.loc[filt,['Age','LanguageWorkedWith','Country']]
```

### HURRAH!!! It worked

#  🤩   🤩   🤩   🤩   🤩

### Now let's try to that for multiple strings

```python
filt = df ['LanguageWorkedWith'].str.contains('Python','C','Java', na = False)
```

###  If at first you don't succeed, try, try again.

#  👻  👻  👻  👻  👻

```python
filt = df ['LanguageWorkedWith'].str.contains(['Python','C','Java'], na = False)
```

### Whaaat!! It didn't work!!!!! Again!!!

#  😮  😮  😮  😮  😮  😮

### Actually, only a single string can be given as argument in contains function
### So that means it cannot be done with multiple strings

### That's not true 

# 🤥  🤥  🤥  🤥  🤥  🤥

### Multiple strings can be filtered using regex(regular expression) 

```python
filt = df ['LanguageWorkedWith'].str.contains('Python|Java', na = False)
```

```python
df.loc[filt,['Age','LanguageWorkedWith','Country']]
```

```python
filt = df ['LanguageWorkedWith'].str.contains('Python|HTML', na = False)
```

```python
df.loc[filt,['Age','LanguageWorkedWith','Country']]
```

### Note: Python|HTML is valid but Python | HTML will not work
### Don't Believe me ? See for yourself

#  🙈   🙈   🙈   🙈   🙈   🙈


```python
filt = df ['LanguageWorkedWith'].str.contains('Python | HTML', na = False)
```

```python
df.loc[filt,['Age','LanguageWorkedWith','Country']]
```
